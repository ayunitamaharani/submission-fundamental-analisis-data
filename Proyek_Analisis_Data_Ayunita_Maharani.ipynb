## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

**Contoh pertanyaan bisnis yang memenuhi seluruh elemen SMART**

***"Faktor apa saja yang memengaruhi penurunan conversion rate pada pengguna aplikasi Android di wilayah Jabodetabek sebesar 5% selama periode Flash Sale Maret 2026?"***

Keterangan:

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.
- **Measurable**: Ada angka konkret yang ingin dianalisis, yaitu penurunan sebesar "5%".
- **Action-Oriented**: Dengan mengetahui faktor penyebabnya misalnya bug pada versi Android tertentu atau kendala logistik di Jabodetabek, tim bisa langsung melakukan perbaikan teknis atau operasional.
- **Relevant**: Penurunan konversi saat Flash Sale adalah masalah kritis bagi bisnis retail/e-commerce.
- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".

## Cleaning Data

### Fixing Data Type & Missing Values in `orders_df`

```python
# Mengonversi kolom waktu dari object menjadi datetime
datetime_columns = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for column in datetime_columns:
    orders_df[column] = pd.to_datetime(orders_df[column])

# Menangani missing values pada tanggal pengiriman
orders_df.dropna(subset=['order_delivered_customer_date'], inplace=True)
print(orders_df.info())
orders_df.head()
```

**Insight:**
- Seluruh kolom waktu pada `orders_df` telah berhasil dikonversi ke tipe data datetime.

- Baris yang tidak memiliki tanggal pengiriman telah dihapus untuk menjaga akurasi perhitungan durasi pengiriman.

### Fixing Missing Values & Category Translation in `products_df`

```python
# Mengisi missing values pada nama kategori produk
products_df["product_category_name"].fillna(value="unknown", inplace=True)

# Menggabungkan dengan tabel translasi kategori ke Bahasa Inggris
products_df = pd.merge(
    left=products_df,
    right=product_category_df,
    how="left",
    left_on="product_category_name",
    right_on="product_category_name"
)

# Jika ada kategori yang tidak punya terjemahan, gunakan nama asli
products_df["product_category_name_english"].fillna(products_df["product_category_name"], inplace=True)

# Menghapus kolom lama dan merapikan kolom baru
products_df.drop(columns=["product_category_name"], inplace=True)
products_df.rename(columns={"product_category_name_english": "product_category_name"}, inplace=True)
products_df.head()
```

**Insight:**
- Nama kategori produk kini telah menggunakan Bahasa Inggris dan nilai kosong telah diisi dengan "unknown".

### Fixing Inaccurate Values in `order_payments_df`

```python
# Memastikan tidak ada payment_value yang bernilai 0 atau negatif
order_payments_df = order_payments_df[order_payments_df["payment_value"] > 0]
print(order_payments_df.describe())
order_payments_df.head()
```

### EDA Pertanyaan 1: Performa Penjualan & Musiman (São Paulo, Nov 2017)

```python
# Revenue São Paulo - Perbandingan Okt vs Nov 2017
sp_df = full_df[full_df['customer_state'] == 'SP'].copy()
sp_df['year_month'] = sp_df['order_purchase_timestamp'].dt.to_period('M')

# Revenue per bulan di SP
sp_monthly_rev = (sp_df.groupby('year_month')['payment_value']
                  .sum().reset_index())
sp_monthly_rev.columns = ['year_month', 'revenue']

# Filter Okt & Nov 2017
sp_okt = sp_monthly_rev[sp_monthly_rev['year_month'] == '2017-10']['revenue'].values[0]
sp_nov = sp_monthly_rev[sp_monthly_rev['year_month'] == '2017-11']['revenue'].values[0]
pct_change = ((sp_nov - sp_okt) / sp_okt) * 100

print(f"Revenue SP Oktober 2017 : R$ {sp_okt:,.2f}")
print(f"Revenue SP November 2017: R$ {sp_nov:,.2f}")
print(f"Perubahan               : {pct_change:+.1f}%")

# Top 10 kategori kontributor di SP November 2017
sp_nov_df = sp_df[sp_df['year_month'] == '2017-11']
top_cat_nov = (sp_nov_df.groupby('product_category_name')['payment_value']
               .sum().sort_values(ascending=False).head(10).reset_index())
top_cat_nov.columns = ['product_category', 'revenue']
top_cat_nov['revenue_pct'] = (top_cat_nov['revenue'] / sp_nov * 100).round(2)

print("\nTop 10 Kategori Produk di SP November 2017:")
print(top_cat_nov.to_string(index=False))
```

**Insight:**
- Revenue São Paulo memang melonjak signifikan di November 2017, bertepatan dengan momen **Black Friday**.
- Beberapa kategori produk tertentu mendominasi kontribusi pendapatan di bulan November dibandingkan Oktober.
- Analisis ini membantu mengidentifikasi kategori "unggulan" yang perlu distok lebih banyak menjelang November berikutnya.

### EDA Pertanyaan 2: Efisiensi Pengiriman Bahia vs Nasional (Q1 2018)

```python
# Waktu Pengiriman Bahia vs Nasional Q1 2018
delivery_df = full_df.copy()
delivery_df['delivery_days'] = (
    delivery_df['order_delivered_customer_date'] -
    delivery_df['order_purchase_timestamp']
).dt.days

q1_2018 = delivery_df[
    (delivery_df['order_purchase_timestamp'] >= '2018-01-01') &
    (delivery_df['order_purchase_timestamp'] < '2018-04-01')
].copy()

# Rata-rata pengiriman: nasional vs Bahia
avg_national = q1_2018['delivery_days'].mean()
avg_bahia    = q1_2018[q1_2018['customer_state'] == 'BA']['delivery_days'].mean()
diff_days    = avg_bahia - avg_national

print(f"Rata-rata pengiriman nasional Q1 2018 : {avg_national:.1f} hari")
print(f"Rata-rata pengiriman ke Bahia Q1 2018 : {avg_bahia:.1f} hari")
print(f"Selisih                               : {diff_days:+.1f} hari")

# Rata-rata review score per state
review_by_state = (q1_2018.groupby('customer_state')['review_score']
                   .mean().sort_values().reset_index())
review_by_state.columns = ['state', 'avg_review_score']

# Review score Bahia vs nasional
review_national = q1_2018['review_score'].mean()
review_bahia    = q1_2018[q1_2018['customer_state'] == 'BA']['review_score'].mean()

print(f"\nRata-rata review score nasional : {review_national:.2f}")
print(f"Rata-rata review score Bahia    : {review_bahia:.2f}")

# Korelasi pengiriman vs review score (per state)
state_stats = (q1_2018.groupby('customer_state')
               .agg(avg_delivery=('delivery_days','mean'),
                    avg_review=('review_score','mean'))
               .reset_index())
corr = state_stats[['avg_delivery','avg_review']].corr().iloc[0,1]
print(f"\nKorelasi delivery_days vs review_score (per state): {corr:.3f}")
print("\nTop 10 State dengan pengiriman terlama Q1 2018:")
print(state_stats.sort_values('avg_delivery', ascending=False).head(10).to_string(index=False))
```

**Insight:**
- Pengiriman ke Bahia (BA) memang lebih lambat dibandingkan rata-rata nasional pada Q1 2018, konsisten dengan hipotesis awal.
- Terdapat korelasi negatif antara lama pengiriman dan review score semakin lambat pengiriman, semakin rendah rating ulasan pelanggan.
- Wilayah dengan pengiriman terlama umumnya berada di North dan Northeast Brazil (seperti AM, PA, MA), yang menunjukkan tantangan infrastruktur logistik di wilayah terpencil.

### EDA Pertanyaan 3: Loyalitas Pelanggan Churn H1 2018

```python
# Customer Loyalty & Churn H1 2018
# Menghitung frekuensi transaksi per unique customer
cust_freq = (full_df.groupby('customer_unique_id')
             .agg(total_orders=('order_id','nunique'),
                  last_purchase=('order_purchase_timestamp','max'))
             .reset_index())

# Klasifikasi: Loyal = transaksi > 2 kali
cust_freq['segment'] = cust_freq['total_orders'].apply(
    lambda x: 'Loyal' if x > 2 else 'Regular'
)

loyal_customers = cust_freq[cust_freq['segment'] == 'Loyal']
print(f"Total unique customers : {len(cust_freq):,}")
print(f"Pelanggan Loyal (>2x)  : {len(loyal_customers):,} ({len(loyal_customers)/len(cust_freq)*100:.1f}%)")

# Definisi churn: loyal customers yang last purchase < 2018-07-01
churn_cutoff = pd.Timestamp('2018-07-01')
loyal_churn = loyal_customers[loyal_customers['last_purchase'] < churn_cutoff]
churn_rate  = len(loyal_churn) / len(loyal_customers) * 100

print(f"\nLoyalitas + Churn H1 2018:")
print(f"Loyal customers churn (last purchase < Jul 2018): {len(loyal_churn):,}")
print(f"Churn rate loyal customers                      : {churn_rate:.1f}%")

# Kategori produk terakhir yang dibeli oleh loyal churn customers
churn_ids = loyal_churn['customer_unique_id'].tolist()

last_orders = (full_df[full_df['customer_unique_id'].isin(churn_ids)]
               .sort_values('order_purchase_timestamp')
               .groupby('customer_unique_id')
               .last()
               .reset_index())

top_last_cat = (last_orders['product_category_name']
                .value_counts().head(10).reset_index())
top_last_cat.columns = ['product_category', 'count']

print("\nTop 10 Kategori Produk Terakhir yang Dibeli Loyal Churn Customers:")
print(top_last_cat.to_string(index=False))
```

**Insight:**
- Hanya sebagian kecil pelanggan yang melakukan transaksi lebih dari 2 kali, menandakan tingkat retensi yang perlu ditingkatkan.
- Mayoritas pelanggan loyal pun terdampak churn, artinya program loyalitas eksisting belum cukup efektif.
- Kategori produk terakhir yang dibeli memberikan sinyal tentang di mana pengalaman pelanggan mungkin kurang memuaskan dan perlu di*remarketing*.

### EDA Pertanyaan 4: Metode Pembayaran vs Tingkat Pembatalan (2017)

```python
# Boleto vs Credit Card - Cancellation Rate 2017 ─────────────
orders_all = pd.read_csv("orders_dataset.csv")
orders_all['order_purchase_timestamp'] = pd.to_datetime(orders_all['order_purchase_timestamp'])

orders_2017 = orders_all[orders_all['order_purchase_timestamp'].dt.year == 2017].copy()
pay_type = order_payments_df[order_payments_df['payment_sequential'] == 1][['order_id','payment_type']]

orders_pay_2017 = pd.merge(orders_2017, pay_type, on='order_id', how='left')

# Filter hanya boleto dan credit_card
orders_pay_2017 = orders_pay_2017[orders_pay_2017['payment_type'].isin(['boleto','credit_card'])]

# Hitung cancellation rate per payment type
cancel_stats = (orders_pay_2017.groupby('payment_type')
                .apply(lambda x: pd.Series({
                    'total_orders'     : len(x),
                    'canceled_orders'  : (x['order_status'] == 'canceled').sum(),
                    'cancellation_rate': (x['order_status'] == 'canceled').mean() * 100
                })).reset_index())

print("Cancellation Rate berdasarkan Metode Pembayaran (2017):")
print(cancel_stats.to_string(index=False))

boleto_rate = cancel_stats[cancel_stats['payment_type']=='boleto']['cancellation_rate'].values[0]
cc_rate     = cancel_stats[cancel_stats['payment_type']=='credit_card']['cancellation_rate'].values[0]
print(f"\nSelisih Boleto vs Credit Card: {boleto_rate - cc_rate:+.2f}%")

# Distribusi status order per payment type
status_dist = (orders_pay_2017.groupby(['payment_type','order_status'])
               .size().reset_index(name='count'))
print("\nDistribusi Status Order per Metode Pembayaran (2017):")
print(status_dist[status_dist['order_status'].isin(['delivered','canceled','shipped'])]
      .to_string(index=False))
```

**Insight:**

## Analisis Lanjutan: RFM Analysis

RFM Analysis digunakan untuk mengelompokkan pelanggan berdasarkan tiga dimensi perilaku:
- **Recency (R)**: Jumlah hari sejak transaksi terakhir
- **Frequency (F)**: Jumlah total transaksi unik
- **Monetary (M)**: Total pengeluaran pelanggan

Segmentasi ini membantu menentukan strategi marketing yang tepat untuk setiap kelompok pelanggan.

```python
rfm_base = full_df[full_df['order_status'] == 'delivered'].copy()
reference_date = rfm_base['order_purchase_timestamp'].max() + pd.Timedelta(days=1)
print(f"Tanggal referensi RFM: {reference_date.date()}")

rfm = (rfm_base.groupby('customer_unique_id')
       .agg(
           recency   = ('order_purchase_timestamp', lambda x: (reference_date - x.max()).days),
           frequency = ('order_id', 'nunique'),
           monetary  = ('payment_value', 'sum')
       ).reset_index())

# Scoring RFM dengan quantile (1-4)
rfm['R_score'] = pd.qcut(rfm['recency'],   q=4, labels=[4,3,2,1]).astype(int)  # lower recency = better
rfm['F_score'] = pd.qcut(rfm['frequency'].rank(method='first'), q=4, labels=[1,2,3,4]).astype(int)
rfm['M_score'] = pd.qcut(rfm['monetary'].rank(method='first'),  q=4, labels=[1,2,3,4]).astype(int)

rfm['RFM_Score'] = rfm['R_score'] + rfm['F_score'] + rfm['M_score']

# Segmentasi berdasarkan RFM Score
def rfm_segment(row):
    if row['R_score'] >= 3 and row['F_score'] >= 3 and row['M_score'] >= 3:
        return 'Champions'
    elif row['R_score'] >= 3 and row['F_score'] >= 2:
        return 'Loyal Customers'
    elif row['R_score'] >= 3 and row['F_score'] <= 2:
        return 'Potential Loyalists'
    elif row['R_score'] == 2:
        return 'At Risk'
    else:
        return 'Lost Customers'

rfm['Segment'] = rfm.apply(rfm_segment, axis=1)

# Ringkasan per segment
rfm_summary = (rfm.groupby('Segment')
               .agg(jumlah_pelanggan=('customer_unique_id','count'),
                    avg_recency=('recency','mean'),
                    avg_frequency=('frequency','mean'),
                    avg_monetary=('monetary','mean'))
               .round(2)
               .sort_values('jumlah_pelanggan', ascending=False))

print("\nRFM Segmentation Summary:")
print(rfm_summary.to_string())

# Visualisasi RFM
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle("Analisis Lanjutan: RFM Customer Segmentation",
             fontsize=14, fontweight='bold', y=1.01)

# Plot 1: Distribusi segmen pelanggan
seg_count = rfm['Segment'].value_counts()
colors_rfm = ['#2ecc71','#3498db','#f1c40f','#e67e22','#e74c3c']
wedges, texts, autotexts = axes[0].pie(
    seg_count, labels=seg_count.index, autopct='%1.1f%%',
    colors=colors_rfm[:len(seg_count)], startangle=90,
    wedgeprops={'edgecolor':'white','linewidth':2}
)
for at in autotexts: at.set_fontsize(9); at.set_fontweight('bold')
axes[0].set_title("Distribusi Segmen Pelanggan\n(RFM Analysis)", fontsize=12, fontweight='bold')

# Plot 2: Avg Monetary per segment
rfm_summary_sorted = rfm_summary.sort_values('avg_monetary', ascending=True)
colors_bar = sns.color_palette("viridis", len(rfm_summary_sorted))
bars = axes[1].barh(rfm_summary_sorted.index, rfm_summary_sorted['avg_monetary'],
                    color=colors_bar, edgecolor='white')
axes[1].set_title("Rata-rata Pengeluaran per Segmen\n(Avg Monetary Value)",
                  fontsize=12, fontweight='bold')
axes[1].set_xlabel("Rata-rata Pengeluaran (R$)")
axes[1].grid(axis='x', alpha=0.3)

for bar, val in zip(bars, rfm_summary_sorted['avg_monetary']):
    axes[1].text(bar.get_width() + 2, bar.get_y() + bar.get_height()/2,
                 f"R${val:,.0f}", va='center', fontsize=9)

plt.tight_layout()
plt.savefig("viz_rfm.png", dpi=150, bbox_inches='tight')
plt.show()

print("\nRFM Analisis & Visualisasi selesai.")
```

**Insight:**
- Distribusi segmen pelanggan cukup merata, namun Lost Customers mendominasi (25.0%) diikuti At Risk (24.9%) yang artinya hampir 50% pelanggan sudah tidak aktif atau berisiko churn.
- Segmen Champions hanya 13.1% dari total pelanggan, namun memiliki avg monetary tertinggi sebesar 378 BRL hampir 3x lipat dibandingkan Loyal Customers 134 BRL, menjadikan mereka segmen paling bernilai.
- Loyal Customers justif memiliki avg monetary terendah (134 BRL) di antara semua segmen, mengindikasikan pelanggan yang sering bertransaksi belum tentu bertransaksi dengan nilai besar.
- Segmen At Risk dengan avg recency 277 hari perlu mendapat perhatian segera krena mereka pernah aktif dan berbelanja cukup besar yakni 216 BRL namun sudah lama tidak bertransaksi.
- Rata-rata frequency hampir seluruh segmen mendekati 1.0, mengkonfirmasi temuan EDA sebelumnya bahwa **mayoritas pelanggan hanya bertransaksi sekali**.
